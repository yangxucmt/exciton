```julia
using CSV, DataFrames, Plots, LaTeXStrings
```

```julia
df = CSV.read("fig_s2.csv", DataFrame)

strain = df.strainlist
rshift_wannier = df.rshift_wannier
rshift_full_2pi_3 = df.rshift_full_2pi_3
rshift_full_3pi_4 = df.rshift_full_3pi_4;
rshift_full_5pi_6 = df.rshift_full_5pi_6;

hline([0], color = :black, lw = 0.5, label = false, xlim = (-6.1, 6.1), ylim = (-0.4, 0.4), xticks = ([-6, -3, 0, 3, 6]))

# Vertical axis at x = 0
vline!([0], color = :black, lw = 0.5, label = false)
scatter!(
	strain,
	[rshift_wannier, rshift_full_2pi_3, rshift_full_3pi_4, rshift_full_5pi_6] .*100,
	framestyle = :box,
	xgrid = false,
	ygrid = false,
	label = ["Wannier" "Full, "*L"\theta=2\pi/3" "Full, "*L" \theta=3\pi/4" "Full, "*L" \theta=5\pi/6"],
	tickfontsize = 12,
	dpi = 800,
	size = (400, 280),
	markershape = [:circle :square :utriangle :diamond],
	color=[:black :orange :blue :red],
	ms = [5.5 4 4 5],
	markeralpha = 0.6,
)

savefig("figs2.pdf")
```
